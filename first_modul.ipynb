```python
df_check = df.copy()
df_check['name_clean'] = df_check['name'].str.lower().str.strip()
df_check['address_clean'] = df_check['address'].str.lower().str.strip()
```

- Созданы нормализованные столбцы: 'name_clean', 'address_clean'

Проверка на явные дубликаты (полные копии строк)

### Проверка на неявные дубликаты (по названию и адресу)

- Дубликаты по названию и адресу

```python
name_address_duplicates = df_check.duplicated(subset=['name_clean', 'address_clean'], keep=False)
duplicates_count = name_address_duplicates.sum()
print(f"Дубликатов по названию и адресу: {duplicates_count}")
```

```python
if duplicates_count > 0:
    print("\nПримеры дубликатов:")
    dup_examples = df_check[name_address_duplicates].sort_values(['name_clean', 'address_clean']).head(10)
    print(dup_examples[['name', 'address', 'category', 'district']].to_string())
else:
    print("Неявных дубликатов не найдено")
```

 ### Проверка на дубликаты по названию (без учёта адреса)

=== ПРОВЕРКА ДУБЛИКАТОВ ПО НАЗВАНИЮ ===

```python
name_duplicates = df_check['name_clean'].duplicated(keep=False).sum()
print(f"Заведения с одинаковыми названиями: {name_duplicates}")

if name_duplicates > 0:
    print("\nПримеры одинаковых названий:")
    name_counts = df_check['name_clean'].value_counts()
    common_names = name_counts[name_counts > 1].head(5)
    
    for name in common_names.index:
        print(f"\nНазвание: '{name}' ({common_names[name]} заведений)")
        examples = df_check[df_check['name_clean'] == name][['name', 'address', 'district']].head(3)
        print(examples.to_string(index=False))
```

=== ПРОВЕРКА ДУБЛИКАТОВ ПО АДРЕСУ ===

```python
address_duplicates = df_check['address_clean'].duplicated(keep=False).sum()
print(f"Адреса с несколькими заведениями: {address_duplicates}")

if address_duplicates > 0:
    print("\nПримеры адресов с несколькими заведениями:")
    address_counts = df_check['address_clean'].value_counts()
    common_addresses = address_counts[address_counts > 1].head(5)
    
    for addr in common_addresses.index:
        print(f"\nАдрес: '{addr}' ({common_addresses[addr]} заведений)")
        examples = df_check[df_check['address_clean'] == addr][['name', 'category', 'district']].head(3)
        print(examples.to_string(index=False))
```

**Вывод:**
- Уделить особое внимание качеству обслуживания в ночные часы
- Ориентироваться на формат быстрого питания с базовым меню

###  Для ценообразования

#### Учитывать районный коэффициент

| Группа районов | Средний чек | Отличие от центра |
|----------------|------------|-------------------|
| Центр | 1 191 ₽ | — |
| Ближние районы (Северный, Западный) | 900–1 050 ₽ | -12–25% |
| Средние районы (Южный, Восточный) | 800–850 ₽ | -28–33% |
| Отдалённые районы | 500–600 ₽ | -50–58% |

**Вывод:**
- В центре можно устанавливать цены на **20–30% выше** среднего по городу
- В спальных районах — **конкурентные цены** для привлечения локальной аудитории
- Удалённость от центра должна коррелировать со стоимостью меню

- Наиболее успешные сети в сегменте кофеен и фастфуда
- При расширении сети учитывать районную специфику цен и формата

##  ЗАКЛЮЧЕНИЕ

Рынок общественного питания Москвы характеризуется:
- **8 406 заведений**, из которых 62% — несетевые, 38% — сетевые
- **Высокая концентрация в центре** (30% всех заведений)
- **Средний рейтинг** — 4.23 (от 1 до 5)
- **Средний чек** — 958 ₽ (медиана 750 ₽)

**Основные драйверы успеха:**
- **Локация** (центр для премиум-сегмента, спальные районы для демократичных форматов)
- **Качество** (кухня, сервис, атмосфера)

**Цены определяются в первую очередь географией, а не форматом заведения.**  
Сети занимают значительную долю рынка (38%), но **не имеют преимущества в рейтингах** перед независимыми заведениями.

Для успешного ведения бизнеса важно сочетать правильный выбор локации с фокусом на качество обслуживания.
